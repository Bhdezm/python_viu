```python
def cat_age(data):
    '''
    Función que imputa el valor -0.5 a los NaN
    Bines: -1, 0, 12, 18, 30, 60, 120
    Input: pandas dataframe
    Output: pandas dataframe
    '''
    data['Age'] = data['Age'].fillna(-0.5)
    bins = (-1, 0, 12, 18, 30, 60, 120)
    cat_names = ['Unknown', 'Child', 'Teenagers', 'Young Adult', 'Adult', 'Elderly']
    categorias = pd.cut(data['Age'], bins, labels=cat_names)
    data['Age'] = categorias
    return
```

```python
cat_age(train_data)
```

```python
cat_age(test_data)
```

## ¿Qué hacemos con los camarotes?

```python
sns.countplot(x='Cabin', data=train_data)
```

```python
def extract_cabin(data):
    '''
    Función para tratar los camarotes
    Gestión de NAs: asignamos la letra N cuando no tengan camarote asignado
    Para el resto de camarotes, nos quedamos con la primera letra del identificador de camarote
    Input: pandas daaframe
    Output: pandas dataframe
    '''
    data['Cabin'] = data['Cabin'].fillna('N')
    data['Cabin'] = data['Cabin'].apply(lambda x: x[0])
    return
```

```python
extract_cabin(train_data)
extract_cabin(test_data)
```

```python
train_data.sample(10)
```

## ¿Qué pasa con los precios?

```python
def cat_fare(data):
    '''
    Función para imputar los valores faltantes de las tarifas y categorizarlos
    Input: pandas dataframe
    Output: pandas dataframe
    '''
    data['Fare'] = data['Fare'].fillna(0.0)
    cat_names = ['1st', '2nd', '3rd', '4th', '5th']
    data['Fare'] = pd.qcut(data['Fare'], 5, labels=cat_names)
    return
```

```python
cat_fare(train_data)
cat_fare(test_data)
```

```python
train_data.sample(10)
```

## ¿Qué hacemos con los nombres?

```python
train_data['Name'].sample(15)
```

```python
def extract_title(data):
    '''
    Función para extraer el título de las personas
    Input: pandas dataframe
    Output: pandas dataframe
    '''
    data['Title'] = data['Name'].apply(lambda x: x.split(',').split()[0])
    return
```

```python
extract_title(train_data)
extract_title(test_data)
```

## Eliminamos que no nos sirven para el análisis

```python
columns_to_drop = ['Name', 'Ticket', 'Embarked']

def drop_columns(data, columns):
    '''
    Función que elimina dado un dataframe eelimina las columnas indicadas en una lista
    Input:
        - data: pandas dataframe
        - columns: list
    Output: pandas dataframe
    '''
    data.drop(columns, axis=1, inplace=True)
    return
```

## Modelando

## Codificar nuestras variables categóricas en forma de números

```python
from sklearn import preprocessing
```

```python
df_train = train_data.copy()
df_test = test_data.copy()
```

```python
def encode_features(df_train, df_test, features):
    '''
    '''
    df_combined = pd.concat([df_train[features], df_test[features]])
    
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(df_combined[feature])
        df_train[feature] = le.transform(df_train[feature])
        df_test[feature] = le.transform(df_test[feature])
        print(f"Feature {feature} OK")
```

```python
features = ['Sex', 'Age', 'Fare', 'Cabin', 'Title']
encode_features(df_train, df_test, features)
df_train.sample(5)
```

```python
df_test.sample(10)
```

## Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
y = df_train['Survived']
X = df_train.drop(['Survived', 'PassengerId'], axis=1)

validation_split = 0.15
seed = 42
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=validation_split, random_state=seed)
```

## Hacemos un modelo

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

preds = model.predict(X_validation)
```

```python
accuracy_score(preds, y_validation)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_validation, preds))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_validation, preds))
```
