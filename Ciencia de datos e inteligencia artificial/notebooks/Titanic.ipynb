```python
def cat_age(data):
    ''' 
    A los que no tengan edades(NaN) se les asigna edad de -0.5
    Bines: -1,0,12,18,30,60,120
    input: pd Dataframe
    output: pd Dataframe
    '''
    data['Age']=data['Age'].fillna(-0.5)#rellenar los NaN
    bins=(-1,0,12,18,30,60,120)
    cat_names=['Unkown','Child','Teenagers','Young Adult','Adult','Elderly']
    categorias=pd.cut(data['Age'],bins,labels=cat_names)
    data['Age']=categorias
    return
```

```python
cat_age(train_data)
```

```python
cat_age(test_data)
```

```python
def cat_fare(data):
    data['Fare']=data['Fare'].fillna(0.0)
    cat_names=['1st','2nd','3rd','4th','5th']
    data['Fare']=pd.qcut(data['Fare'],5,labels=cat_names)
    return
```

## Modelando

## Codificar nuestras variables categóricas en forma de números

```python
from sklearn import preprocessing
```

```python
df_train=train_data.copy()
```

```python
df_test=test_data.copy()
```

```python
def encode_features(df_train,df_test,features):
    ''' '''
    df_combined=pd.concat([df_train[features],df_test[features]])
    
    for feature in features:
        le=preprocessing.LabelEncoder()
        le=le.fit(df_combined[feature])
        df_train[feature]=le.transform(df_train[feature])
        df_test[feature]=le.transform(df_test[feature])
        print(f'Feature {feature} OK')
```

```python
features=['Sex','Age','Fare','Cabin','Title']
encode_features(df_train,df_test,features)
df_train.sample(5)
```

## Train split

```python
from sklearn.model_selection import train_test_split
```

```python
y=df_train['Survived']
X=df_train.drop(['Survived','PassengerId'],axis=1)
validation_split=0.15
seed=42
X_train,X_validation,y_train,y_validation=train_test_split(X,y,test_size=validation_split,random_state=seed)
```

## Hacemos un modelo

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

preds=model.predict(X_validation)
```

```python
accuracy_score(preds,y_validation)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_validation,preds))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_validation,preds))
```
